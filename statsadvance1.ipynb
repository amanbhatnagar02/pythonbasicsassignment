#question 3:Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
import numpy as np

def bernoulli_pdf(p, x):
    """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 ≤ p ≤ 1).
    x (int): Random variable (either 0 or 1).

    Returns:
    float: Probability P(X = x)
    """
    if x not in [0, 1]:  # Bernoulli random variable can only be 0 or 1
        return 0
    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage:
p = 0.7  # Probability of success
x_values = [0, 1]  # Possible values of Bernoulli variable

# Compute PDF for both possible values
pdf_values = [bernoulli_pdf(p, x) for x in x_values]

# Display results
for x, pdf in zip(x_values, pdf_values):
    print(f"P(X={x}) = {pdf}")
```

#question 6: > Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define parameters for the discrete uniform distribution
low = 1  # Minimum value
high = 6  # Maximum value (inclusive)
size = 10000  # Number of samples

# Generate discrete uniform distributed data
data = np.random.randint(low, high + 1, size)

# Sort data to compute CDF
x_values, counts = np.unique(data, return_counts=True)
cdf_values = np.cumsum(counts) / size  # Compute cumulative probability

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x_values, cdf_values, where='mid', color='blue', label="CDF", linewidth=2)
plt.scatter(x_values, cdf_values, color='red', zorder=3)  # Mark points
plt.xlabel("Random Variable (X)")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Distribution")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()
plt.show()
```

#question 9: Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculates Z-scores for a dataset and plots a histogram.

    Parameters:
    - data: A list or numpy array of numerical values
    """
    # Convert to numpy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7, color='skyblue')
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.axvline(0, color='red', linestyle='dashed', label='Mean (Z=0)')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores

# ✅ Add this part to make it work
data = np.random.normal(loc=50, scale=10, size=1000)
z_scores = calculate_and_plot_zscores(data)
```

#question 14:  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    z = (data_point - mean) / std_dev
    return z

def plot_z_score(z):
    # Generate x values for standard normal curve
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Curve', color='blue')

    # Fill area under curve up to the Z-score
    x_fill = np.linspace(-4, z, 1000)
    y_fill = norm.pdf(x_fill, 0, 1)
    plt.fill_between(x_fill, y_fill, color='orange', alpha=0.6, label=f'Area ≤ Z ({z:.2f})')

    # Mark Z-score with a vertical line
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Z-Score on Standard Normal Distribution')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data_point = 85
mean = 70
std_dev = 10

z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point}: {z:.2f}")

plot_z_score(z)
```

#question 15 : Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

# Population parameters
population_mean = 75
population_std_dev = 10

# Sample data
sample_mean = 78
sample_size = 50

# Step 1: Calculate Standard Error
standard_error = population_std_dev / np.sqrt(sample_size)

# Step 2: Calculate Z-statistic
z_score = (sample_mean - population_mean) / standard_error

# Step 3: Calculate p-value (for one-tailed test)
p_value = 1 - norm.cdf(z_score)

# Step 4: Set significance level
alpha = 0.05

# Step 5: Decision
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print(" Reject the null hypothesis: The sample mean is significantly higher.")
else:
    print(" Fail to reject the null hypothesis: Not enough evidence.")
```

#Question 16: 5 Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy import stats

# Step 1: Sample data (can be any dataset)
sample_data = [72, 75, 78, 74, 77, 79, 80, 73, 76, 74]

# Step 2: Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample std dev
sample_size = len(sample_data)

# Step 3: Set confidence level
confidence = 0.95
alpha = 1 - confidence

# Step 4: Find the t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)

# Step 5: Calculate margin of error
margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))

# Step 6: Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

#Question 17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
from scipy import stats

# Step 1: Generate random data from normal distribution
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=50)  # Mean=100, SD=15, Sample size=50

# Step 2: Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation
n = len(data)

# Step 3: Calculate the confidence interval
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # t-distribution

# Step 4: Margin of error
margin_error = t_critical * (sample_std / np.sqrt(n))

# Step 5: Confidence interval
lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

#Question 23 : 5 Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)

# Step 2: Z-score normalization
mean = np.mean(data)
std_dev = np.std(data)
z_scores = (data - mean) / std_dev

# Step 3: Plot original and normalized data
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()
plt.grid(True)

# Z-score normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black')
plt.title('Z-Score Normalized Data')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Optional: Print verification
print(f"Original Mean: {mean:.2f}, Std Dev: {std_dev:.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")
```

#Question 24:Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution(size=1000, mean=0, std=1):
    """
    Generate data from a normal distribution and calculate skewness and kurtosis.

    Parameters:
    - size: Number of data points
    - mean: Mean of the distribution
    - std: Standard deviation

    Returns:
    - skew_val: Skewness of the data
    - kurtosis_val: Kurtosis of the data (excess kurtosis by default)
    """
    # Generate data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurtosis_val = kurtosis(data)  # excess kurtosis (normal = 0)

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis (Excess): {kurtosis_val:.4f}")

    return skew_val, kurtosis_val

# Example usage
analyze_normal_distribution()
```
