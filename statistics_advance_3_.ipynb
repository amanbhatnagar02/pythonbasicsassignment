#Question 9: How do you calculate the Z-score, and what does it represent in hypothesis testing

ans: To calculate the Z-score, you do this:

Find how much your sample result differs from the average (subtract the population mean from your sample result).

Divide that difference by the standard deviation (which tells you the typical amount of variation).

So, the Z-score =

“Difference between sample and average” ÷ “How much things usually vary”

If you’re testing a group (not just one value), you also divide by the square root of the number of observations to adjust for sample size.

#Question 18: What is the Chi-square goodness of fit test, and how is it applied

The Chi-square goodness of fit test checks whether the observed frequencies of categories in your data match the expected frequencies from a specific theoretical distribution or hypothesis.

In other words, it helps you decide if your data fits well with what you expect.

How is it applied?
Collect your observed data — these are the actual counts or frequencies you get from your experiment or survey for each category.

Decide the expected counts — based on your hypothesis or theory, figure out how many counts you expect in each category if the theory is true.

Calculate the difference between observed and expected counts for each category.

Square the differences to make sure all are positive and to give more weight to bigger differences.

Divide each squared difference by the expected count — this standardizes the differences based on the size of what you expected.

Add up all these values across all categories to get the Chi-square statistic.


#Question 19:  What is the F-distribution, and when is it used in hypothesis testing

The F-distribution is mainly used in analysis of variance (ANOVA) tests, which compare the variances (or means) of three or more groups to see if at least one group mean is different.

It is also used to test the equality of variances between two populations (F-test for variances).

In regression analysis, the F-test checks whether the overall regression model is a good fit for the data.

Normality: The data in each group are approximately normally distributed.

Homogeneity of variances (Homoscedasticity): The variances across groups are roughly equal.

#Question :  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
from scipy.stats import norm

# --- Step 1: Input your known values ---
population_mean = 100        # Known population mean
population_std = 15          # Known population standard deviation
sample_data = [105, 98, 110, 102, 95, 108, 99, 100, 103, 97]  # Sample data

# --- Step 2: Calculate sample statistics ---
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# --- Step 3: Compute Z-statistic ---
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# --- Step 4: Compute p-value (two-tailed) ---
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# --- Step 5: Print results ---
print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# --- Step 6: Interpretation ---
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print("Interpretation: The sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("Interpretation: There is no significant difference between the sample and population means.")
```

#Question 5 :Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    z_critical = norm.ppf(1 - alpha / 2)
    se = sigma / np.sqrt(n)

    x = np.linspace(mu0 - 4*se, mu0 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    lower_bound = mu0 - z_critical * se
    upper_bound = mu0 + z_critical * se

    plt.figure(figsize=(12, 6))

    plt.plot(x, y0, label='H0: Mean = μ0', color='blue')
    plt.plot(x, y1, label='H1: Mean = μ1', color='green')

    plt.fill_between(x, y0, where=(x <= lower_bound) | (x >= upper_bound), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, y1, where=(x > lower_bound) & (x < upper_bound), color='purple', alpha=0.3, label='Type II Error (β)')

    plt.axvline(lower_bound, color='black', linestyle='--')
    plt.axvline(upper_bound, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
visualize_type1_type2_errors(mu0=100, mu1=103, sigma=15, n=30, alpha=0.05)
```

#Question 9: Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2, df=n-1)
    lower_bound = mean - margin
    upper_bound = mean + margin
    return (lower_bound, upper_bound)

    np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)

ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval for the mean: {ci}")
```

#question 10: Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    n = len(data)
    sem = stats.sem(data)  # Standard Error of the Mean
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_score * sem
    return moe

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=12, size=40)

confidence_level = 0.95
moe = margin_of_error(sample_data, confidence=confidence_level)

print(f"Margin of Error at {int(confidence_level * 100)}% confidence level: {moe:.2f}")

```

#Question 11:  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(prior_H, prob_E_given_H, prob_E_given_not_H):
    prob_not_H = 1 - prior_H
    prob_E = (prob_E_given_H * prior_H) + (prob_E_given_not_H * prob_not_H)
    posterior = (prob_E_given_H * prior_H) / prob_E
    return posterior

# Parameters
prior_disease = 0.01            # P(H): 1% prior probability of disease
prob_positive_given_disease = 0.95   # P(E|H): 95% chance test is positive if diseased
prob_positive_given_no_disease = 0.05 # P(E|~H): 5% false positive rate

# Calculate posterior
posterior = bayes_theorem(prior_disease, prob_positive_given_disease, prob_positive_given_no_disease)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

#Question 13:Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

```python
import pandas as pd
import numpy as np

def calculate_expected_frequencies(observed):
    row_totals = observed.sum(axis=1).values.reshape(-1, 1)
    col_totals = observed.sum(axis=0).values.reshape(1, -1)
    grand_total = observed.values.sum()
    expected = (row_totals @ col_totals) / grand_total
    expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
    return expected_df

observed_data = pd.DataFrame({
    'Low': [30, 20],
    'Moderate': [10, 40],
    'High': [5, 50]
}, index=['Smoker', 'Non-Smoker'])

print("Observed Frequencies:")
print(observed_data)

expected_frequencies = calculate_expected_frequencies(observed_data)

print("\nExpected Frequencies:")
print(expected_frequencies)

```

#Question 14:  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (e.g., counts in categories)
observed = np.array([50, 30, 20])

# Expected frequencies (must sum to same total as observed)
expected = np.array([40, 40, 20])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject null hypothesis: Observed data fits the expected distribution.")

```

#Question 16:  Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=100, scale=10, size=30)
sample2 = np.random.normal(loc=100, scale=15, size=30)

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Ensure larger variance is in numerator
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(sample1) - 1, len(sample2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(sample2) - 1, len(sample1) - 1

# Calculate the p-value (two-tailed test)
p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

print(f"Sample 1 Variance: {var1:.4f}")
print(f"Sample 2 Variance: {var2:.4f}")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

```python
import numpy as np
import scipy.stats as stats

def check_anova_assumptions(*groups):
    print("=== Checking ANOVA Assumptions ===")

    group_labels = [f"Group {i+1}" for i in range(len(groups))]

    print("\nNormality (Shapiro-Wilk Test):")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"{group_labels[i]}: W={stat:.4f}, p={p:.4f} → {'Normal' if p > 0.05 else 'Not Normal'}")

    print("\nEqual Variance (Levene’s Test):")
    stat, p = stats.levene(*groups)
    print(f"Levene’s Statistic={stat:.4f}, p={p:.4f} → {'Equal variances' if p > 0.05 else 'Unequal variances'}")

    print("\nIndependence is assumed based on study design (not tested statistically).")

np.random.seed(42)
group1 = np.random.normal(loc=65, scale=8, size=30)
group2 = np.random.normal(loc=70, scale=8, size=30)
group3 = np.random.normal(loc=75, scale=8, size=30)

check_anova_assumptions(group1, group2, group3)

```

#Question 20 :  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

np.random.seed(42)
data = pd.DataFrame({
    'Treatment': np.repeat(['A', 'B'], 40),
    'Gender': np.tile(np.repeat(['Male', 'Female'], 20), 2),
    'Score': np.concatenate([
        np.random.normal(70, 5, 20),
        np.random.normal(75, 5, 20),
        np.random.normal(80, 5, 20),
        np.random.normal(85, 5, 20)
    ])
})

model = ols('Score ~ C(Treatment) * C(Gender)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("Two-Way ANOVA Results:")
print(anova_table)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Treatment', y='Score', hue='Gender', data=data, palette='Set2')
plt.title('Score by Treatment and Gender')
plt.ylabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()

from statsmodels.graphics.factorplots import interaction_plot
plt.figure(figsize=(8, 6))
interaction_plot(data['Treatment'], data['Gender'], data['Score'],
                 colors=['red', 'blue'], markers=['D', '^'], ms=10)
plt.title('Interaction Plot: Treatment x Gender')
plt.ylabel('Mean Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#Question 21: Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

dfn_values = [1, 5, 10]  # Numerator degrees of freedom
dfd_values = [5, 10, 20]  # Denominator degrees of freedom
x = np.linspace(0, 5, 500)

plt.figure(figsize=(10, 6))
for dfn, dfd in zip(dfn_values, dfd_values):
    y = f.pdf(x, dfn, dfd)
    plt.plot(x, y, label=f'dfn={dfn}, dfd={dfd}')

plt.title('F-distribution PDFs for Different Degrees of Freedom')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

#Question 22:  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means

#Question 23:  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

```python
import numpy as np
from scipy.stats import ttest_ind

# Simulate data from two normal distributions
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=40)  # mean=50, std=5
sample2 = np.random.normal(loc=52, scale=5, size=40)  # mean=52, std=5

# Perform two-sample t-test (independent samples)
t_stat, p_value = ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between means.")
```

#Question 24: Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy.stats import chi2

np.random.seed(42)
sample = np.random.normal(loc=50, scale=7, size=30)

s2 = np.var(sample, ddof=1)

sigma0_squared = 49

n = len(sample)

chi_square_stat = (n - 1) * s2 / sigma0_squared

df = n - 1

alpha = 0.05

lower_crit = chi2.ppf(alpha / 2, df)
upper_crit = chi2.ppf(1 - alpha / 2, df)

print(f"Sample variance: {s2:.4f}")
print(f"Chi-square statistic: {chi_square_stat:.4f}")
print(f"Critical values: Lower={lower_crit:.4f}, Upper={upper_crit:.4f}")

if chi_square_stat < lower_crit or chi_square_stat > upper_crit:
    print("Reject the null hypothesis: population variance differs from hypothesized value.")
else:
    print("Fail to reject the null hypothesis: insufficient evidence to say population variance differs.")
```

#Question 26: Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)

# Simulate two datasets
data1 = np.random.normal(loc=50, scale=5, size=40)
data2 = np.random.normal(loc=50, scale=8, size=40)

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# F-test statistic (ratio of variances, larger variance numerator)
if var1 > var2:
    f_stat = var1 / var2
    dfn = len(data1) - 1  # numerator degrees of freedom
    dfd = len(data2) - 1  # denominator degrees of freedom
else:
    f_stat = var2 / var1
    dfn = len(data2) - 1
    dfd = len(data1) - 1

# Calculate p-value for two-tailed test
p_value = 2 * min(stats.f.cdf(f_stat, dfn, dfd), 1 - stats.f.cdf(f_stat, dfn, dfd))

print(f"Variance of data1: {var1:.4f}")
print(f"Variance of data2: {var2:.4f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: variances are significantly different.")
else:
    print("Fail to reject null hypothesis: no significant difference in variances.")

# Visualization
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.hist(data1, bins=15, alpha=0.7, label='Data 1 (var={:.2f})'.format(var1))
plt.hist(data2, bins=15, alpha=0.7, label='Data 2 (var={:.2f})'.format(var2))
plt.legend()
plt.title('Histogram of Both Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1,2,2)
plt.boxplot([data1, data2], labels=['Data 1', 'Data 2'])
plt.title('Boxplot Comparison')
plt.ylabel('Value')

plt.tight_layout()
plt.show()
```

#Question 27:  Perform a Chi-square test for goodness of fit with simulated data and analyze the results

```python
import numpy as np
from scipy.stats import chisquare

# Simulated observed frequencies (e.g., counts of categories)
observed = np.array([50, 30, 20, 40])

# Expected frequencies assuming a uniform distribution
expected = np.full_like(observed, np.mean(observed))

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: observed distribution differs from expected.")
else:
    print("Fail to reject the null hypothesis: no significant difference from expected distribution.")
```
